## Création des fonctions

```python
def get_match_data(region, match_id, api_key):
    api_url = "https://{}.api.riotgames.com/lol/match/v5/matches/{}?api_key={}".format(region, match_id, api_key)
    response = requests.get(api_url)
    data = response.json()
    return data
```

```python
match_data = get_match_data(region_api[2], matchlist[0], api_key)
```

```python
match_data
```

```python
match_data["info"]["participants"]
```

```python
def did_win(puuid, match_data):
    part_index = match_data["metadata"]["participants"].index(puuid)
    return match_data["info"]["participants"][part_index]["win"]
```

```python
for index, match_id in enumerate(matchlist):
    if index >= 20:
        break
    match_data = get_match_data(region_api[2], match_id, api_key)
    print(did_win(puuid, match_data))
```
